```python
import requests
from bs4 import BeautifulSoup
from time import sleep
from urllib.parse import urljoin
import os

def get_pokemon_file(page_url, pokemon_name):

    # Send a GET request to the page URL
    page_response = requests.get(page_url)
    sleep(1)
    # Check if the page request was successful
    if page_response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(page_response.text, 'html.parser')
        
        # Find all <td> elements with colspan="4"
        td_elements = soup.find_all('td', {'colspan': '4'})
        
        # Initialize variable to store the final link
        final_link = None

        # Iterate through the found <td> elements to find the <a> with class "image"
        for td in td_elements:
            a_image = td.find('a', class_='image')
            if a_image:
                # Check if the <a> element contains an <img> with the specified width and height
                img = a_image.find('img', width="250", height="250")
                if img:
                    final_link = a_image['href']
                    break
        
        if final_link:
            # Handle potential relative URL
            if final_link.startswith('//'):
                final_link = 'https:' + final_link
            elif final_link.startswith('/'):
                # Construct full URL if it's a relative path
                from urllib.parse import urljoin
                final_link = urljoin(page_url, final_link)
            
            # Follow the final_link to get to the download page
            download_page_response = requests.get(final_link)
            if download_page_response.status_code == 200:
                soup_download_page = BeautifulSoup(download_page_response.text, 'html.parser')
                # Find the <a> element with the exact text
                download_link = soup_download_page.find('a', string="download the original file here", class_="external text")
                
                if download_link and 'href' in download_link.attrs:
                    download_href = download_link['href']
                    # Handle potential relative URL for the download link
                    if download_href.startswith('//'):
                        download_href = 'https:' + download_href
                    
                    # Download the file
                    file_response = requests.get(download_href, stream=True)
                    
                    if file_response.status_code == 200:
                        with open(f"images/pokemon/{pokemon_name}.png", 'wb') as file:
                            for chunk in file_response.iter_content(chunk_size=128):
                                file.write(chunk)
                        print(f"File downloaded successfully and saved as {pokemon_name}.png")
                    else:
                        print("Failed to download the file from the download link.")
                else:
                    print("Download link not found on the final page.")
            else:
                print("Failed to retrieve the download page.")
        else:
            print("Specific link not found.")
    else:
        print("Failed to retrieve the initial page.")
```

```python
def get_move_file(page_url, move_name):
    # Check if the file already exists
    file_path = f"images/pokemon_moves/{move_name}.png"
    if os.path.exists(file_path):
        # print(f"The file {move_name}.png already exists. Skipping download.")
        return
    print(f"Downloading {move_name} from {page_url}")
    # Send a GET request to the page URL, which is now directly the final link
    page_response = requests.get(page_url)
    sleep(1)  # Wait a bit before parsing the page, to be polite to the server

    # Check if the page request was successful
    if page_response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(page_response.text, 'html.parser')
        
        # Find the <a> element with the exact text "download the original file here" and class "external text"
        download_link = soup.find('a', string="download the original file here", class_="external text")

        if download_link and 'href' in download_link.attrs:
            download_href = download_link['href']
            # Handle potential relative URL for the download link
            if download_href.startswith('//'):
                download_href = 'https:' + download_href
            elif download_href.startswith('/'):
                download_href = urljoin(page_url, download_href)

            # Attempt to download the file
            file_response = requests.get(download_href, stream=True)

            if file_response.status_code == 200:
                with open(file_path, 'wb') as file:
                    for chunk in file_response.iter_content(chunk_size=128):
                        file.write(chunk)
                print(f"File downloaded successfully and saved as {move_name}.png")
            else:
                print("Failed to download the file from the download link.")
        else:
            print("Download link not found on the page.")
    else:
        print("Failed to retrieve the page.", page_url)
        
```
