```python
def apply_chen_formula(hand, data):
    score = 0

    '''
    Score your highest card only. Do not add any points for your lower card.
    A = 10 points.
    K = 8 points.
    Q = 7 points.
    J = 6 points.
    10 to 2 = 1/2 of card value. (e.g. a 6 would be worth 3 points)
    '''
    highest_card = hand[0]
    if highest_card == "A":
        score += 10
    elif highest_card == "K":
        score += 8
    elif highest_card == "Q":
        score += 7
    elif highest_card == "J":
        score += 6
    else:
        score += data["high_value"] / 2

    '''
    Multiply pairs by 2 of one cardâ€™s value. However, minimum score for a pair is 5.
    (e.g. KK = 16 points, 77 = 7 points, 22 = 5 points)
    '''
    if hand[0] == hand[1]:
        score = max(5, score * 2) 

    '''
    Add 2 points if cards are suited.
    '''
    if hand[2] == "s":
        score += 2

    '''
    Subtract points if there is a difference between the two cards.
    No difference = -0 points.
    1 card gap = -1 points.
    2 card gap = -2 points.
    3 card gap = -4 points.
    4 card gap or more = -5 points. (Aces are high this step, so hands like A2, A3 etc. have a 4+ gap.)
    '''
    difference = data["difference"]
    if difference <= 0:
        pass
    elif difference == 1:
        score -= 1
    elif difference == 2:
        score -= 2
    elif difference == 3:
        score -= 4
    else:
        score -= 5
    '''
    Add 1 point if there is a 0 or 1 card gap and both cards are lower than a Q. (e.g. JT, 75, 32 etc, this bonus point does not apply to pocket pairs)
    '''
    if difference <= 1 and hand[0] != hand[1]:
        if data["high_value"] < 12 and data["low_value"] < 12:
            score += 1

    '''
    Round half point scores up. (e.g. 7.5 rounds up to 8)
    '''
    score = round(score)
    return score
```

```python
handlist = raw_handlist.split('\n')[1:-1]
handlist = [x.split('\t') for x in handlist[1:]]

hands_dict = {card[1] : {"mrank" : int(card[0])} for card in handlist}
cards_value_transposition_values = {
    "A" : 14,
    "K" : 13,
    "Q" : 12,
    "J" : 11,
    "T" : 10,
    "9" : 9,
    "8" : 8,
    "7" : 7,
    "6" : 6,
    "5" : 5,
    "4" : 4,
    "3" : 3,
    "2" : 2,
}

for hand, data in hands_dict.items():
    data["high_value"] = cards_value_transposition_values[hand[0]]
    data["low_value"] = cards_value_transposition_values[hand[1]]
    data["sum_of_cards"] = data["high_value"] + data["low_value"]
    data["suited"] = 1 if hand[2] == "s" else 0
    data["difference"] = abs(data["high_value"] - data["low_value"])
    data["gap"] = data["difference"] - 1
    data["is_pair"] = 1 if data["difference"] == 0 else 0
    data["chen_score"] = apply_chen_formula(hand, data)

chen_scores = [data["chen_score"] for hand, data in hands_dict.items()]
chen_scores.sort(reverse=True)

chen_rank_differences = []
for hand, data in hands_dict.items():
    data["chen_rank"] = chen_scores.index(data["chen_score"]) + 1
    data["chen_rank_difference"] = abs(data["chen_rank"] - data["mrank"])
    chen_rank_differences.append(data["chen_rank_difference"])

print(sum(chen_rank_differences) / len(chen_rank_differences))
print(statistics.median(chen_rank_differences))


chen_squared_differences = [x**2 for x in chen_rank_differences]
chen_mean_squared_difference = sum(chen_squared_differences)/len(chen_squared_differences)
chen_root_mean_squared_difference = math.sqrt(chen_mean_squared_difference)

print(f"RMSE: {chen_root_mean_squared_difference}")
# median_difference = statistics.median(differences)
```

```python
def get_custom_score(a,b,c,d,e=0):
    test_scores = []
    hand_data_sheet = {}

    for hand, data in hands_dict.items():
                    
        chen_1 = 0
        highest_card = hand[0]
        if highest_card == "A": chen_1 += 10
        elif highest_card == "K": chen_1 += 8
        elif highest_card == "Q": chen_1 += 7
        elif highest_card == "J": chen_1 += 6
        else: chen_1 += data["high_value"] / 2

        score =  data["suited"]*a + data["difference"]*b + data["is_pair"]*c 
        score += data["sum_of_cards"]*d 
        score += chen_1*e

        if data["mrank"] <= 20 or data["mrank"] >= 165:
            score = 170-data["mrank"]

        test_scores.append(score)
        hand_data_sheet[hand] = {
            "cards": hand,
            "rank": data["mrank"],
            "score": score,
            "chen_score": data["chen_score"],
            "chen_rank": data["chen_rank"],
        }
    return (test_scores, hand_data_sheet)

def get_normalized_test_differences(test_scores, hand_data_sheet):
    test_differences = []
    test_scores.sort(reverse=True)

    for hand, data in hand_data_sheet.items():
        data["custom_test_normalized_rank"] = test_scores.index(data["score"]) + 1
        data["custom_test_difference"] = abs(data["custom_test_normalized_rank"] - data["rank"])
        test_differences.append(data["custom_test_difference"])

    return (test_differences, hand_data_sheet)


def test_custom_scoring_algorithm(a,b,c,d,e=0, to_print=False):
    (test_scores, hand_data_sheet) = get_custom_score(a,b,c,d,e)

    (test_differences, hand_data_sheet) = get_normalized_test_differences(test_scores, hand_data_sheet)

    scores = [round(data["score"]) for hand, data in hand_data_sheet.items()]
    if to_print == True:
        # print(scores)
        # print(len(scores))
        print(f"Preflop Hand\tBrute Force Rank\tCustom Score\tNormalized Custom Score (Rank)\tDiff")
        for hand, data in hand_data_sheet.items():
            pass
            
            print(f"{hand}\t{data['rank']}\t{170-round(data['score'])}\t{data['custom_test_normalized_rank']}\t{data['custom_test_difference']}")


    # biggest_difference = max(differences)
    mean_difference = sum(test_differences)/len(test_differences)

    squared_differences = [x**2 for x in test_differences]
    mean_squared_difference = sum(squared_differences)/len(squared_differences)
    root_mean_squared_difference = math.sqrt(mean_squared_difference)
    average_score_delta = abs(85 - abs(sum(scores)/len(scores)))

    if to_print == True:
        print(f"RMSE: {root_mean_squared_difference}")
        print(f"Mean Difference: {mean_difference}")
        print(f"Average Score Delta: {average_score_delta}")
    # median_difference = statistics.median(differences)

    average_score_delta = abs(85 - abs(sum(scores)/len(scores)))

    return mean_difference * (1 + average_score_delta)
```

```python
weights = [6.2, 0.3, 64.2, 4.0, 1.7]

def upper(num):
    return num + 100.0
def lower(num):
    return num - 100.0

# print(test_custom_scoring_algorithm(*weights, to_print=True))
bestscore = test_custom_scoring_algorithm(*weights, to_print=True)
print(bestscore)
```

```python
# in this loop structure, we want to fine-tune each weight before we move on.
while True:
    changed = False
    for i, weight in enumerate(weights):
        # print(f"testing weight {i}")
        for j in range(int(upper(weight)*10), int(lower(weight)*10), -1):
            test_weights = [*weights[:i], j/10, *weights[i+1:]]
            score = test_custom_scoring_algorithm(*test_weights)
            # print(score)
            if score < bestscore:
                changed = True
                bestscore = score
                print(f"New best score: {bestscore}")
                # print(f"New best weights: {test_weights}")
                weights = test_weights
            # else:
                # weights[i] = bestscore
        # print(i)
    if changed == False:
        break
print(weights, bestscore)
```
